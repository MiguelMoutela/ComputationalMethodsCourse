### Derivatives

Formula for the derivative of a Fourier-transormed function:

$$\hat{f^{(n)}} = (ik)^n \hat{f}$$

```python
sech = lambda x: 1./(np.cosh(x))
u = sech(x)
ut = fft(u)
# True first derivative
ud = -sech(x)*tanh(x)
# True second derivative
ud2 = sech(x)-2*sech(x)**3
```

```python
plot(x, ud)
```

```python
# Estimated first derivative
ud_spect = ifft((1j*k)*ut)
# Estiamted second derivative
ud2_spect = ifft((1j*k)**2 * ut)
```

```python
plot(x, ud)
plot(x,ud_spect,'r--')
```

```python
plot(x, ud2)
plot(x, ud2_spect, 'r--')
```

```python
T = 30
n = 512

t = linspace(-T/2., T/2., n, endpoint=False)
```

```python
k = (2*pi/T)*r_[linspace(0, n/2-1, n/2), linspace(-n/2,-1,n/2)]
```

```python
# Signal as function of time
u = sech(t)
ut = fft(u)
```

```python
subplot(211)
plot(t,u)
subplot(212)
plot(fftshift(k), abs(fftshift(ut))/max(abs(fftshift(ut))))
xlim(-15,15)
```

Now, add noise to every frequency component (white noise)

```python
noise = 10
utn = ut + noise*(randn(n) + 1j*randn(n))
# Transform back to time domain
un = ifft(utn)
```

```python
subplot(211)
plot(t,abs(un),t,u,'r--')
subplot(212)
plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))),fftshift(k), abs(fftshift(ut))/max(abs(fftshift(ut))),'r--')
xlim(-15,15)
```

### Denoising

Can we retrieve the signal from the above?

```python
filter_ = exp(-k**2)

plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))),fftshift(k), fftshift(filter_), 'r--')
```

Multiply the noisy signal by the filter

```python
utnf = filter_ * utn

plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))))
plot(fftshift(k), fftshift(filter_), 'r--')
plot(fftshift(k), abs(fftshift(utnf))/max(abs(fftshift(utnf))), 'k--')
xlim(-5,5)
```

```python
unf = ifft(utnf)

plot(t,abs(un),alpha=0.2)
plot(t,abs(unf),'k--',t,[0.5]*len(t),'r')
```

Shift the filter.

```python
filter_ = exp(-(k+15)**2)

utnf = filter_ * utn

unf = ifft(utnf)

subplot(211)
plot(t,abs(un),alpha=0.2)
plot(t,abs(unf),'k--',t,[0.5]*len(t),'r')
subplot(212)
plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))))
plot(fftshift(k), fftshift(filter_), 'r--')
plot(fftshift(k), abs(fftshift(utnf))/max(abs(fftshift(utnf))), 'k--')
xlim(-20,20)
```

## Lecture 3

Averaging over white noise in signals.

```python
# Sample for T seconds
T = 30
# Sample with n points
n = 512
```

```python
t = linspace(-T/2., T/2., n, endpoint=False)
k = (2*pi/T)*r_[linspace(0, n/2-1, n/2), linspace(-n/2,-1,n/2)]
```

```python
u = sech(t)
ut = fft(u)
```

```python
noise = 30
utn = ut + noise*(randn(n) + 1j*randn(n))
un = ifft(utn)
```

```python
subplot(211)
plot(t,abs(un),'k',t,u,'r--')
subplot(212)
plot(fftshift(k), abs(fftshift(utn)),fftshift(k), abs(fftshift(ut)),'r--')
#xlim(-15,15)
```

Sample this data repeatedly, with changing noise for each sample. Then average the samples, which should remove the zero-mean noise.

```python
realizations = 30
samples = np.sum([ut + noise*(randn(n) + 1j*randn(n)) for i in range(realizations)], 0)
avg = abs(fftshift(samples))/realizations

plot(fftshift(k), abs(fftshift(ut)), 'r--')
plot(fftshift(k), avg, 'k')
plot(fftshift(k), abs(fftshift(utn)), alpha=0.3)
```

We can animate this averaging process:

```python
import time, sys
from IPython.display import clear_output
f, ax = plt.subplots()

samples = zeros(n)
for i in range(realizations):
    samples += ut + noise*(randn(n) + 1j*randn(n))
    avg = abs(fftshift(samples))/realizations
    ax.plot(fftshift(k), abs(fftshift(ut)), 'r--', alpha=0.3)
    ax.plot(fftshift(k), avg, 'k')
    time.sleep(0.5)
    clear_output()
    display(f)
    ax.cla() 
plt.close()
```

This will give us an idea about where to filter.

### Moving signals

Moving signals will be lost if we average in the time domain. We need to exploit the frequency domain.

```python
T = 60

t = linspace(-T/2., T/2., n, endpoint=False)
k = (2*pi/T)*r_[linspace(0, n/2-1, n/2), linspace(-n/2,-1,n/2)]

timeslice = arange(0, 10.5, 0.5)
T,S = meshgrid(t, timeslice)
K,S = meshgrid(k, timeslice)
```

```python
U = sech(T - 10*sin(S)) * exp(1j*0*T)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import PolyCollection
from matplotlib.colors import colorConverter

fig = plt.figure()
ax = fig.gca(projection='3d')

poly = PolyCollection([zip(i,j) for i,j in zip(T,U)])
poly.set_alpha(0.7)
ax.add_collection3d(poly, zs=timeslice, zdir='y')

ax.set_ylim3d(0, 10)
ax.set_xlim3d(T.min()*.6, T.max()*.6)
ax.set_zlim3d(0, np.real(U).max()*2)
ax.view_init(20,-120)
```

```python
UT = empty(U.shape)
for i in range(len(timeslice)):
    UT[i] = abs(fftshift(fft(U[i])))
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

poly = PolyCollection([zip(i,j) for i,j in zip(fftshift(K),UT)])
poly.set_alpha(0.7)
ax.add_collection3d(poly, zs=timeslice, zdir='y')

ax.set_ylim3d(0, 10)
ax.set_xlim3d(-10, 10)
ax.set_zlim3d(0, UT.max()*2)
ax.view_init(20,-120)
```

## Lecture 7

Gabor transform and wavelets.

```python
L = 10
n = 2048
t = linspace(0, L, n, endpoint=False)
```

```python
k = (2*pi/L)*r_[linspace(0, n/2-1, n/2), linspace(-n/2, -1, n/2)]
ks = ifftshift(k)
```

The signal.

```python
S = (3*sin(2*t) + 0.5*tanh(0.5*(t-3.)) + 0.28*exp(-(t-4.)**2) + 
     1.5*sin(5.*t) + 4*cos(3*(t-6)**2))/10 + (t/20)**3
```

```python
plot(t,S)
```

Fourier transform of signal:

```python
St = fft(S)
subplot(211)
plot(t,S)
subplot(212)
St_shift = abs(fftshift(St))
plot(ks, St_shift/max(St_shift))
xlim(-80, 80)
```

We want to look at time and frequency together. 

Build a Gaussian filter:

```python
widths = 10, 1, 0.2
f = [exp(-w*(t-4)**2) for w in widths]

subplot(311)
plot(t,S,t,f[0])
subplot(312)
plot(t,S,t,f[1])
subplot(313)
plot(t,S,t,f[2])
```

In the plots above, we are going to multiply the Gaussian filter by the signal.

Lets animate the application of the filter to the signal:

```python
import time, sys
from IPython.display import clear_output

w = 10.
slide = arange(0, 10, step=0.1)

f, ax = subplots(3)
spec = []
for x in slide:
    time.sleep(0.5)
    fx = exp(-w*(t-x)**2)
    # Signal filter
    Sf = fx*S
    # Energy content
    Sft = fft(Sf)
    
    # Plot series
    ax[0].plot(t, S, t, fx)
    ax[1].plot(t, Sf)
    Sf_shift = abs(fftshift(Sft))/max(abs(fftshift(Sft)))
    ax[2].plot(ks, Sf_shift)
    spec.append(Sf_shift)
    xlim(-60,60)
    clear_output()
    display(f)
    ax[0].cla()
    ax[1].cla()
    ax[2].cla()

# close the figure at the end, so we don't get a duplicate
# of the last plot
plt.close()
```
