## Lecture 9

Image processing

```python
photo = imread('data/women_with_guns.png')
imshow(photo)
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray

photo_bw = rgb2gray(photo)
imshow(photo_bw)
```

```python
pcolor(photo_bw[::-1], cmap="Reds")
```

Fourier transform of image.

```python
photo_t = fftshift(fft2(photo_bw))
imshow(log(abs(photo_t)))
#pcolor(log(abs(fftshift(photo_t[::-1]))), cmap="cool")
```

Add noise to image.

```python
photo_noise = photo_bw + random.randn(*shape(photo_bw))*0.3
```

```python
imshow(photo_noise, cmap="gray")
```

Fourier transform noisy image.

```python
photo_noiset = fftshift(fft2(photo_noise))
imshow(abs(photo_noiset))
```

## Lecture 10

Linear filtering of images.

Create filter for noisy image. Try a gaussian in the center of the image.

```python
x_coord = arange(photo_noiset.shape[0])
y_coord = arange(photo_noiset.shape[1])
kx,ky = meshgrid(x_coord, y_coord)
```

```python
F = exp(-0.001*(kx - 257)**2 - 0.001*(ky - 361)**2).T
pcolor(F, cmap="cool")
```

```python
photo_noiset_filtered = photo_noiset * F
photo_filtered = ifft2(fftshift(photo_noiset_filtered))
pcolor(photo_filtered[::-1], cmap="Greys")
xlim(0,700)
ylim(0,500)
```
